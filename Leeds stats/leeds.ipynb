```python
#Busco ver la relación entre los goles por partido y la cantidad de remate al arco

atacantes["Goles/partido"]
```

```python
#Elimino el primer index porque hay un error (es irreal que un jugador tenga 14 goles por partido)

atacantes=atacantes.drop([0])
```

```python
atacantes["Remates eficaces"]
```

```python
#Para analizar necesito convertir los porcentajes a int, para eso primero saco el signo %

lista_remates=[]
for remate in atacantes["Remates eficaces"]:
    remate=remate.replace("%","")
    lista_remates.append(remate)
```

```python
lista_remates
```

```python
#Reemplazo los porcentajes por los números sin porcentaje

atacantes["Remates eficaces"]=lista_remates
```

```python
#Convierto la columna de remates a int para seguir con el análisis

atacantes["Remates eficaces"]=atacantes["Remates eficaces"].astype(int)
```

```python
#Visualizo la tabla de atacantes solo con remates y goles por partido


atacantes[['Goles/partido', 'Remates eficaces']].groupby(['Goles/partido'], as_index=False).mean().sort_values(by='Remates eficaces', ascending=False)
```

```python
#Grafico los datos y veo la relación

plt.bar(atacantes["Remates eficaces"],atacantes["Goles/partido"],label="Relación entre goles y remates",width=2)
plt.title("Relación entre goles y remates")
plt.ylabel("Goles/partido")
plt.xlabel("Remates eficaces")
```
